```python
import numpy as np
import matplotlib.pyplot as plt
```

## Insertion Sort

```python
def insertion_sort(A, n):
    
    for j in range(1, n):
        chave = A[j]
        i = j-1
        while(i>=0 and A[i]>chave):
            A[i+1] = A[i]
            i = i-1
        A[i+1] = chave
        
    return A
```

```python
menor = 0
maior = 10
tam = 10

A = np.random.randint(menor, maior+1, tam)
print('Vetor: ', A)
```

```python
n = len(A)

A_ord = insertion_sort(A, n)
```

```python
print('Vetor ordenado: ', A_ord)
```

## Melhor caso, pior caso e caso médio

Insertion sort com custo computacional (instruções executadas).

```python
def insertion_sort_custo(A, n):
    
    global c
    
    c+=1
    for j in range(1, n):
        c+=1
        chave = A[j]
        c+=1
        i = j-1
        c+=1
        while(i>=0 and A[i]>chave):
            c+=1
            A[i+1] = A[i]
            c+=1
            i = i-1
        c+=1
        A[i+1] = chave
    
    c+=1
    return A
```

```python
n_min = 0
n_max = 1000
n_step = 10

n_array = np.arange(n_min, n_max+n_step, n_step, dtype=np.int64)

c_melhor = []
c_medio = []
c_pior = []

for n in n_array:
    
    A = np.random.randint(0, 1001, n)
    
    A_melhor = sorted(A, reverse=False)
    A_pior = sorted(A, reverse=True)
    A_medio = A.copy() # O caso médio é a própria geração aleatória
    
    c=0
    _ = insertion_sort_custo(A_melhor, n)
    c_melhor.append(c)
    
    c=0
    _ = insertion_sort_custo(A_medio, n)
    c_medio.append(c)
    
    c=0
    _ = insertion_sort_custo(A_pior, n)
    c_pior.append(c)
```

```python
plt.plot(n_array, c_melhor)
plt.ylabel('Instruções executadas')
plt.xlabel('Tamanho da entrada')
plt.title('Entrada ordenada')
plt.grid()
plt.show()
```

```python
plt.plot(n_array, c_pior)
plt.ylabel('Instruções executadas')
plt.xlabel('Tamanho da entrada')
plt.title('Entrada ordenada reversa')
plt.grid()
plt.show()
```

```python
plt.plot(n_array, c_medio)
plt.ylabel('Instruções executadas')
plt.xlabel('Tamanho da entrada')
plt.title('Entrada aleatória')
plt.grid()
plt.show()
```
