```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_min = 0
n_max = 4
n_step = 1

n = np.arange(n_min, n_max+n_step, n_step, dtype=np.int64)
print('Tamanho da entrada: ', n)
```

```python
f = np.power(n, 2) + 2*n + 1
g = np.power(2, n)

plt.plot(n, f)
plt.plot(n, g)
plt.ylabel('Instruções executadas')
plt.xlabel('Tamanho da entrada')
plt.legend(['f(n)', 'g(n)'])
plt.grid()
plt.show()
```
