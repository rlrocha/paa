```python
import numpy as np
import matplotlib.pyplot as plt
import time
import math
import random
import itertools
from collections import defaultdict
```

## Funções para utilização de grafos por lista de adjacências e algortimo dijsktra

```python
class GrafoLista:

    def __init__(self):
        self.nodes = set()
        self.edges = defaultdict(list)
        self.distances = {}

    def adicionar_vertice(self, value):
        self.nodes.add(value)

    def adicionar_aresta(
        self,
        from_node,
        to_node,
        distance,
        ):
        self.edges[from_node].append(to_node)
        self.edges[to_node].append(from_node)
        self.distances[(from_node, to_node)] = distance
        self.distances[(to_node, from_node)] = distance  # Coloquei isso aqui (Deixar direcionado)


def dijsktra_lista(graph, initial):
    visited = {initial: 0}
    path = {}

    nodes = set(graph.nodes)

    while nodes:
        min_node = None
        for node in nodes:
            if node in visited:
                if min_node is None:
                    min_node = node
                elif visited[node] < visited[min_node]:
                    min_node = node

        if min_node is None:
            break

        nodes.remove(min_node)
        current_weight = visited[min_node]

        for edge in graph.edges[min_node]:
            try:
                weight = current_weight + graph.distances[(min_node,
                        edge)]
            except:
                weight = current_weight + math.inf

            if edge not in visited or weight < visited[edge]:
                visited[edge] = weight
                path[edge] = min_node

    return (visited, path)
```

## Funções para utilização de grafos por matriz de adjacências e algortimo dijsktra

```python
class GrafoMatriz:
    # number of vertices
    __n = 0
  
    # adjacency matrix
    __g =[[0 for x in range(10000)] for y in range(10000)] # 10
       
    # constructor
    def __init__(self, x):
        self.__n = x
  
        # initializing each element of the adjacency matrix to zero
        for i in range(0, self.__n):
            for j in range(0, self.__n):
                self.__g[i][j]= 0
    
    def obterMatriz(self):
        
        matriz = np.zeros((self.__n, self.__n), dtype=np.int8)
        
        for i in range(0, self.__n):
            for j in range(0, self.__n):
                matriz[i][j] = self.__g[i][j]
                
        return matriz
    
    def displayAdjacencyMatrix(self):
        print("\n\n Matriz de Adjacências:", end ="")
         
        # displaying the 2D array
        for i in range(0, self.__n):
            print()
            for j in range(0, self.__n):
                print("", self.__g[i][j], end ="")
    
    def adicionar_aresta(self, x, y, w):
        # checks if the vertex exists in the graph
        if(x>= self.__n) or (y >= self.__n):
            print("Vértice não existe!")
          
        # checks if the vertex is connecting to itself
        if(x == y):
             print("Mesmo vértice!")
        else:
               
             # connecting the vertices
             self.__g[y][x]= w
             self.__g[x][y]= w 
 
    def adicionar_vertice(self):
        # increasing the number of vertices
        self.__n = self.__n + 1;
           
        # initializing the new elements to 0
        for i in range(0, self.__n):
            self.__g[i][self.__n-1]= 0
            self.__g[self.__n-1][i]= 0                 
     
    def remover_vertice(self, x):
        
        # checking if the vertex is present
        if(x>self.__n):
             print("Vértice não está presente!")
        else:
            # removing the vertex
            while(x<self.__n):
                
                # shifting the rows to left side
                for i in range(0, self.__n):
                    self.__g[i][x]= self.__g[i][x + 1]
             
                 # shifting the columns upwards
                for i in range(0, self.__n):
                    self.__g[x][i]= self.__g[x + 1][i]
                x = x + 1
  
            # decreasing the number of vertices
            self.__n = self.__n - 1
```

```python
def dijkstra_matriz(graph, start, vis=False):
    
    # This contains the distances from the start node to all other nodes
    distances = [float("inf") for _ in range(len(graph))]

    # This contains whether a node was already visited
    visited = [False for _ in range(len(graph))]

    # The distance from the start node to itself is of course 0
    distances[start] = 0

    # While there are nodes left to visit...
    while True:

        # ... find the node with the currently shortest distance from the start node...
        shortest_distance = float("inf")
        shortest_index = -1
        for i in range(len(graph)):
            # ... by going through all nodes that haven't been visited yet
            if distances[i] < shortest_distance and not visited[i]:
                shortest_distance = distances[i]
                shortest_index = i

        # print("Visiting node " + str(shortest_index) + " with current distance " + str(shortest_distance))

        if shortest_index == -1:
            # There was no node not yet visited --> We are done
            if vis:
                print("Menores distâncias: " + str(distances))
            return distances

        # ...then, for all neighboring nodes that haven't been visited yet....
        for i in range(len(graph[shortest_index])):
            # ...if the path over this edge is shorter...
            if graph[shortest_index][i] != 0 and distances[i] > distances[shortest_index] + graph[shortest_index][i]:
                # ...Save this path as new shortest path.
                distances[i] = distances[shortest_index] + graph[shortest_index][i]
                # print("Updating distance of node " + str(i) + " to " + str(distances[i]))

        # Lastly, note that we are finished with this node.
        visited[shortest_index] = True
        #print("Visited nodes: " + str(visited))
        #print("Currently lowest distances: " + str(distances))
```

## Funções para obter o tempo de execução do algoritmo dijsktra para grafos com lista e matriz de adjacências

### Função auxiliar utilizada para selecionar as arestas a serem inseridas nos grafos

```python
def selecionar_arestas(comb_arestas):
    
    arestas = []
    i=0
    for aresta in comb_arestas:
        if (aresta[0] == i) and (aresta[1] == i+1):
            #print(aresta)
            comb_arestas.remove((i,i+1))
            arestas.append(aresta)
            i+=1
    
    return arestas, comb_arestas
```

### Função principal

```python
def lista_matriz_dijsktra(x_lista, p=1.5, vis=False):
    
    """
    Tempo de execução do algoritmo dijsktra para grafos com lista e matriz de adjacências
    
    
    Parâmetros:
        x_lista (list): Lista de quantidades de vértices dos grafos.
        p (float): Valor utilizado par gerar a quantidade de arestas.
        vis (bool): Visualização do andamento do algoritmo.

    Retornos:
        T_listaDijsktra (list): Tempo obtido pelo algoritmo com lista de adjacências.
        T_matrizDijsktra (list): Tempo obtido pelo algoritmo com matriz de adjacências.

    """
    
    T_listaDijsktra = []
    T_matrizDijsktra = []

    for x in x_lista:

        y = int(x*p) # Número de arestas

        arestas = list(np.arange(0, x, 1))

        comb_arestas = list(itertools.combinations(arestas, 2))

        arestas, comb_arestas = selecionar_arestas(comb_arestas)

        random.shuffle(comb_arestas)

        comb_arestas = comb_arestas[0:y-len(arestas)]

        arestas = arestas + comb_arestas

        # Definição da Lista e da Matriz de Adjacências
        lista_adj = GrafoLista()
        matriz_adj = GrafoMatriz(x)

        # Adicionando os vértices (Lista de adjacências)
        for i in range(x):
            lista_adj.adicionar_vertice(i)

        for aresta in arestas:

            w = np.random.randint(1, 20+1)

            lista_adj.adicionar_aresta(aresta[0], aresta[1], w)
            matriz_adj.adicionar_aresta(aresta[0], aresta[1], w)


        start_time = time.time()
        _ = dijsktra_lista(lista_adj, 0)
        t_lista = time.time() - start_time

        T_listaDijsktra.append(t_lista)

        matriz = matriz_adj.obterMatriz()

        start_time = time.time()
        _ = dijkstra_matriz(matriz, 0)
        t_matriz = time.time() - start_time

        T_matrizDijsktra.append(t_matriz)
        
        # x: Número de vértices
        # y: Número de arestas
        # L: Tempo (s) de execução do algoritmo dijsktra para lista de adjacências
        # M: Tempo (s) de execução do algoritmo dijsktra para matriz de adjacências
        if vis:
            print('x = %d, y = %d: L = %f, M = %f'% (x, y, t_lista, t_matriz))
    
    return T_listaDijsktra, T_matrizDijsktra
```

## Análise da Complexidade dos Algoritmos

Ao analisar o algoritmo dijkstra para a lista e para a matriz de adjacências, nota-se que a matriz de adjacências possui uma maior complexidade em relação ao tempo de execução do algoritmo dijkstra, independentemente do número de arestas, comparado a lista de adjacências. Analisando somente a lista de adjacências, nota-se que a que possui um número de arestas *3x* (*p=3*) maior que o número de vértices é menos eficiente que a que tem arestas *1,5x* (*p=1,5*) o número de vértices. Porém, analisando a matriz de adjacências, nota-se que para *p=3* é ligeiramente mais eficiente que *p=1,5*, indicando uma eficiência melhor quanto mais arestas tem no grafo por matriz de adjacências.

```python
x_lista = np.arange(250, 2250, 250)

T_listaDijsktra1, T_matrizDijsktra1 = lista_matriz_dijsktra(x_lista, 1.5, vis=True)
```

```python
T_listaDijsktra2, T_matrizDijsktra2 = lista_matriz_dijsktra(x_lista, 3, vis=True)
```

```python
fig_luz, axes = plt.subplots(1,2, figsize=(8,4), dpi=300)
plt.tight_layout()

axes[0].plot(x_lista, T_listaDijsktra1, color='blue')
axes[0].plot(x_lista, T_listaDijsktra2, color='red')
axes[0].set_xlabel('Número de vértices')
axes[0].set_ylabel('Tempo de execução (s)')
axes[0].set_title('Lista de adjacências')
axes[0].grid()
axes[0].legend(['p=1.5', 'p=3'])

axes[1].plot(x_lista, T_matrizDijsktra1, color='blue')
axes[1].plot(x_lista, T_matrizDijsktra2, color='red')
axes[1].set_xlabel('Número de vértices')
axes[1].set_ylabel('Tempo de execução (ms)')
axes[1].set_title('Matriz de adjacências')
axes[1].grid()
axes[1].legend(['p=1.5', 'p=3'])

plt.subplots_adjust(right=1.5)

plt.savefig('resultados/ep7/ep7_analise.png', dpi=100, bbox_inches = 'tight')
plt.show()
```
