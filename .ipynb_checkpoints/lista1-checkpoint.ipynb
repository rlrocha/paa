```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
def bubblesort(x):
    
    """
    
    Parameters
    ----------
    x : 
        Vetor de números a serem ordenados.
    
    Returns
    -------
    y :
        Vetor de números ordenados.
    C :
        Quantidade de comandos executados
    T :
        Tempo de execução do algoritmo
    """
        
    C = np.zeros(4) # Comandos executados
    n = np.size(x)

    y = x.copy()
    
    start_time = time.time()
    for i in range(n):
        C[0]+=1 # Custo loop 1
        for j in range(n-1):
            C[1]+=1 # Custo loop 2
            if y[j]>y[j+1]:
                C[2]+=1 # Custo if
                temp = y[j]
                y[j] = y[j+1]
                y[j+1] = temp
                C[3]+=3 # Custo troca
    T = time.time() - start_time # Tempo de execução
    
    return y, C.sum(), T
```

```python
df = pd.read_csv('dados/lista1/n_20.csv')
```

```python
x = df['medio'].values
```

```python
y, C, T = bubblesort(x)
```

```python
print('Vetor: ', x)
print('Vetor ordenado: ', y)
print('Comandos executados: ', np.sum(C))
print('Tempo de execução (ms): ', T*1000)
```

```python
n_list = [10, 20, 30, 40, 50, 60, 70, 80, 100]

C_matriz = np.zeros((len(n_list), 3))
T_matriz = np.zeros((len(n_list), 3))

i=0
for n in n_list:
    nome = 'dados/lista1/n_'+str(n)+'.csv'
    df = pd.read_csv(nome)
    
    melhor = df['melhor'].values
    medio = df['medio'].values
    pior = df['pior'].values
    
    caso_list = [melhor, medio, pior]
    
    for j in range(len(caso_list)):
        
        x, C, T = bubblesort(caso_list[j])
        
        C_matriz[i,j] = C
        T_matriz[i,j] = T
    
    i+=1
```

```python
x_axis = np.arange(1, np.size(C_matriz, 0)+1)
caso = ['Melhor caso', 'Caso médio', 'Pior caso']

plt.plot(x_axis, C_matriz[:,0])
plt.plot(x_axis, C_matriz[:,1])
plt.plot(x_axis, C_matriz[:,2])
plt.legend(caso)
plt.ylabel('Comandos executados')
plt.xlabel('Tamanho da entrada')
plt.title('Bubble sort')
plt.grid()
```

```python
x_axis = np.arange(1, np.size(T_matriz, 0)+1)
caso = ['Melhor caso', 'Caso médio', 'Pior caso']

plt.plot(x_axis, T_matriz[:,0]*1000)
plt.plot(x_axis, T_matriz[:,1]*1000)
plt.plot(x_axis, T_matriz[:,2]*1000)
plt.legend(caso)
plt.title('Bubble sort')
plt.ylabel('Tempo de execução (ms)')
plt.xlabel('Tamanho da entrada')
plt.grid()
```
